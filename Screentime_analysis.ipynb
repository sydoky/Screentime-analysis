```python
import pandas as pd
import numpy as np
import plotly.express as px
import plotly.graph_objects as go

data=pd.read_csv("/content/Screentime-App-Details.csv")
```

```python
data.head(15)
```

```python
data.isnull().sum()
```

# **Let's anaylize the amount of usage of the apps**

```python
figure=px.bar(data_frame=data,x="Date", y="Usage",color="App",title="Usage of Apps")
figure.show()
```

# **Let's analyze number of nofitications from the Apps**

```python
figure=px.bar(data_frame=data,x="Date",y="Notifications",color="App",title="Number of notifications from Apps")
figure.show()
```

# **Analyzing times App opened**

```python
figure=px.bar(data_frame=data,x="Date", y="Times opened",color="App",title="Analyzing how many time Apps were opened")
figure.show()
```
